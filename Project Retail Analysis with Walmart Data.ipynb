```python
#importing relevent libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#importing data
walmart=pd.read_csv('Walmart_Store_sales.csv')
```

```python
#Store with maximum sales
walmart.groupby('Store')['Weekly_Sales'].sum().to_frame().sort_values(by='Weekly_Sales',ascending=False).head(1)
#store with maximum sales is '20'
```

```python
#Store with max standard deviation i.e., the sales vary a lot. And, finding the coefficient of mean to standard deviation
walmart.groupby('Store')['Weekly_Sales'].std().to_frame().reset_index().sort_values(by='Weekly_Sales',ascending=False).head(1)
#Store with maximum Standard deviation in terms of sales is '14'
```

```python
#Saving top store with maximum standard deviation in a dataframe
Store_14_max_dev=walmart.groupby('Store')['Weekly_Sales'].std().to_frame().reset_index().sort_values(by='Weekly_Sales',ascending=False).head(1).reset_index(drop=True)
```

```python
Store_14_max_dev
```

```python
#Mean of weekly sales of store 14
mean_store_14=walmart[walmart['Store']==14]['Weekly_Sales'].mean()
```

```python
mean_store_14
```

```python
#coefficient of mean to standard deviation of store 14
#formula{(standard deviation/mean)*100}
round(((Store_14_max_dev.iloc[0,1]/mean_store_14)*100),2)
```

```python
#Some holidays have a negative impact on sales.
#Finding out holidays which have higher sales than the mean sales in non-holiday season for all stores together

#mean sales of non holiday season
Mean_sales_non_holiday_season=round((walmart[walmart['Holiday_Flag']==0]['Weekly_Sales'].mean()),2)
```

```python
#Saving a dataframe where holidays season sales are higher sales than the mean of non-holiday season sales for all stores
wal_holiday=walmart[(walmart['Holiday_Flag']==1) & (walmart['Weekly_Sales']>Mean_sales_non_holiday_season)]
wal_holiday
```

```python
#holidays which have higher sales than the mean sales in non-holiday season for all stores together
high_sales_holiday=list(wal_holiday['Date'])
high_sales_holiday
```

```python
#store(s) which have good quarterly growth rate in Q3’2012

#adding a quarter number column
walmart['quarter']=pd.to_datetime(walmart['Date']).dt.quarter
#adding a year column
walmart['year']=pd.to_datetime(walmart['Date']).dt.year

```

```python
#quarter 3 sales storewise
wal_q3_sales=walmart[(walmart['year']==2012) & (walmart['quarter']==3)].groupby('Store')['Weekly_Sales'].sum().to_frame().reset_index()
#quarter 2 sales storewise
wal_q2_sales=walmart[(walmart['year']==2012) & (walmart['quarter']==2)].groupby('Store')['Weekly_Sales'].sum().to_frame().reset_index()
#renaming columns
wal_q3_sales.rename(columns={'Weekly_Sales':'total q3 sales 2012'},inplace=True)
wal_q2_sales.rename(columns={'Weekly_Sales':'total q2 sales 2012'},inplace=True)
```

```python
wal_q2_sales.head()
```

```python
wal_q3_sales.head()
```

```python
#merging wal_q3_sales and wal_q2_sales
walmart_q2_q3_sales=pd.merge(wal_q2_sales,wal_q3_sales,on='Store')
walmart_q2_q3_sales.head()
```

```python
# growth rate in Q3 2012= ((Q3 growth rate-Q2 growth rate)/Q2 growth rate)*100
walmart_q2_q3_sales['q3 growth rate 2012']=((walmart_q2_q3_sales['total q3 sales 2012']-walmart_q2_q3_sales['total q2 sales 2012'])/walmart_q2_q3_sales['total q2 sales 2012'])*100
```

```python
walmart_q2_q3_sales.sort_values(by='q3 growth rate 2012',ascending=False).head(3)
# Store 16 has good Q3 growth rate compared to the the rest of the stores in 2012
```

```python
#Building prediction model for store 1 to forecast demand
#{x:features,y:label}
x=walmart[walmart['Store']==1].drop(['Fuel_Price','Date','Weekly_Sales'],axis=1)# no significant relation of sales with fuel price
y=walmart[walmart['Store']==1]['Weekly_Sales']
```

```python
#splittinng training data and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=40)
```

```python
#selecting model
model=LinearRegression()
```

```python
#training model
model.fit(x_train,y_train)
```

```python
#predicting sales
y_pred=model.predict(x_test)
list(y_pred)
```
